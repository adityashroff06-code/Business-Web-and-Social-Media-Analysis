```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset (replace with actual data)
data = {
    "name": ["Cereal A", "Cereal B", "Cereal C", "Cereal D", "Cereal E"],
    "type": ["Cold", "Hot", "Cold", "Cold", "Hot"],
    "calories": [110, 90, 120, 100, 80],
    "protein": [2, 3, 2, 3, 4],
    "fat": [1, 0, 2, 1, 0],
    "sodium": [180, 150, 200, 170, 100],
    "fiber": [2, 4, 1, 3, 5],
    "carbo": [14, 16, 12, 15, 18],
    "sugars": [7, 5, 9, 6, 3],
    "potass": [90, 100, 80, 120, 130],
    "vitamins": [25, 25, 0, 25, 100],
    "weight": [1, 1, 1, 1, 1],
    "cups": [0.75, 1, 0.67, 1, 1],
    "rating": [50.2, 60.5, 40.3, 55.1, 70.2],
    "shelf": [3, 1, 2, 3, 1]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# (a) Identify variable types
quantitative_vars = ['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars', 'potass', 'vitamins', 'weight', 'cups', 'rating']
ordinal_vars = ['shelf']
nominal_vars = ['name', 'type']

# (b) Compute summary statistics
summary_stats = df[quantitative_vars].describe().T
print("\nSummary Statistics:\n", summary_stats)

# (c) Histogram Analysis using Matplotlib (Instead of Seaborn)
plt.figure(figsize=(12, 8))
for i, col in enumerate(quantitative_vars, 1):
    plt.subplot(4, 3, i)
    plt.hist(df[col], bins=5, color="skyblue", edgecolor="black")
    plt.title(col)
plt.tight_layout()
plt.show()

# (d) Boxplots using Matplotlib (Instead of Seaborn)
plt.figure(figsize=(6, 4))
df.boxplot(column="calories", by="type")
plt.title("Calories in Hot vs. Cold Cereals")
plt.suptitle("")
plt.xlabel("Cereal Type")
plt.ylabel("Calories")
plt.show()

plt.figure(figsize=(6, 4))
df.boxplot(column="rating", by="shelf")
plt.title("Consumer Rating by Shelf Height")
plt.suptitle("")
plt.xlabel("Shelf Height (1=bottom, 2=middle, 3=top)")
plt.ylabel("Consumer Rating")
plt.show()

# (e) Correlation Matrix
corr_matrix = df[quantitative_vars].corr()
print("\nCorrelation Matrix:\n", corr_matrix)
```
