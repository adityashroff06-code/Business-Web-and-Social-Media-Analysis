```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the dataset
file_path = "winecsv.csv"  # Change path if necessary
wine_data = pd.read_csv(file_path)

# Drop the categorical 'Wine' column (assumed to be the class label)
X = wine_data.drop(columns=["Wine"], errors='ignore')

# Perform PCA without normalization
pca = PCA()
pca.fit(X)
explained_variance_ratio = pca.explained_variance_ratio_

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA on standardized data
pca_scaled = PCA()
pca_scaled.fit(X_scaled)
explained_variance_ratio_scaled = pca_scaled.explained_variance_ratio_

# Plot explained variance for both cases
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.bar(range(1, len(explained_variance_ratio)+1), explained_variance_ratio*100)
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained (%)')
plt.title('PCA Without Normalization')

plt.subplot(1,2,2)
plt.bar(range(1, len(explained_variance_ratio_scaled)+1), explained_variance_ratio_scaled*100)
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained (%)')
plt.title('PCA With Normalization')

plt.tight_layout()
plt.show()

# Print variance explained by first few components
print("Explained Variance Without Normalization:", explained_variance_ratio[:5])
print("Explained Variance With Normalization:", explained_variance_ratio_scaled[:5])
```
