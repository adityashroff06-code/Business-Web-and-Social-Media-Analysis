```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/LaptopSales.csv'
df = pd.read_csv("/content/LaptopSales.csv")

# Ensure the Date column is in datetime format with the correct format specification
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y %H:%M', errors='coerce')

# Part A: Price Questions

# i. At what price are the laptops actually selling? (Boxplot for overall price)
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Retail Price'])
plt.title('Boxplot of Retail Price for All Laptops')
plt.xlabel('Retail Price')
plt.show()

```

```python
# ii. Does price change with time? (Line graph showing price over time)
# Resample data by month and calculate the average retail price per month
df_monthly = df.resample('M', on='Date')['Retail Price'].mean()

plt.figure(figsize=(10, 6))
df_monthly.plot(kind='line', color='b')
plt.title('Average Retail Price by Month')
plt.xlabel('Month')
plt.ylabel('Average Retail Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# iii. Are prices consistent across retail outlets? (Boxplot by store)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Store Postcode', y='Retail Price', data=df)
plt.title('Boxplot of Retail Price by Store')
plt.xlabel('Store')
plt.ylabel('Retail Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# iv. How does price change with configuration? (Side by side box plot by configuration)
plt.figure(figsize=(12, 6))
sns.boxplot(x='Configuration', y='Retail Price', data=df)
plt.title('Boxplot of Retail Price by Configuration')
plt.xlabel('Configuration')
plt.ylabel('Retail Price')
plt.xticks(rotation=90)
plt.show()
```
