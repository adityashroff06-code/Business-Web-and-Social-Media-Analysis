```python
# prompt: Using Python, create a scatter plot of Lot Size vs. Income, color-coded by the outcome variable owner/nonowner. Make sure to obtain a well-formatted plot (create legible labels and alegend,etc.).

import pandas as pd
import matplotlib.pyplot as plt

# Assuming your data is in a CSV file named 'data.csv'
# Replace 'data.csv' with the actual file name if different
try:
  data = pd.read_csv('/content/RidingMowers (1).csv')
except FileNotFoundError:
  print("Error: 'RidingMowers.csv' not found. Please make sure the file exists and the name is correct.")
  exit()

# Check if required columns exist
required_columns = ['Lot Size', 'Income', 'Ownership']
if not all(col in data.columns for col in required_columns):
    print("Error: The DataFrame does not contain the required columns ('Lot Size', 'Income', 'Ownership').")
    exit()


# Create the scatter plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed

# Assuming 'Owner' column contains 'owner' or 'nonowner' strings
# If your outcome variable is represented differently, adjust the following lines
for owner_status in data['Ownership'].unique():
  subset = data[data['Ownership'] == owner_status]
  plt.scatter(subset['Lot Size'], subset['Income'], label=owner_status, alpha=0.7)  # Adjust alpha for point transparency

# Customize the plot
plt.title('Lot Size vs. Income (Color-coded by Ownership Status)', fontsize=14)
plt.xlabel('Lot Size', fontsize=12)
plt.ylabel('Income', fontsize=12)
plt.legend(title='Ownership Status', fontsize=10) # Add legend
plt.grid(True, linestyle='--', alpha=0.5)  # Add gridlines
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```
