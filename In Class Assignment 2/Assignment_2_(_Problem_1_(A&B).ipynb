```python
# prompt: Shipments of Household Appliances: Line Graphs. The file ApplianceShipments.csv contains the series of quarterly shipments (in millions of dollars) of US household appliances between 1985 and 1989.
# a. Create a well-formatted time plot of the data using Python.

import pandas as pd
import matplotlib.pyplot as plt

# Install pandas and matplotlib if not already installed
!pip install pandas matplotlib

# Load the data
try:
  df = pd.read_csv('/content/ApplianceShipments.csv')
except FileNotFoundError:
  print("Error: ApplianceShipments.csv not found. Please upload the file.")
  # You can use google.colab.files.upload() here if needed.
  # Example:
  # from google.colab import files
  # uploaded = files.upload()
  # df = pd.read_csv(list(uploaded.keys())[0])
  exit()

# Convert the 'Dates' column to datetime objects, handling different formats
# If 'Dates' contains just the year, assume it's the first quarter
# Convert 'Dates' column to string type first
df['Dates'] = df['Dates'].astype(str)
df['Quarter'] = pd.to_datetime(df['Dates'].str.replace(r'^(\d{4})$', r'\1-01-01', regex=True), format='%Y-%m-%d').dt.to_period('Q').dt.start_time


# Create the time plot
plt.figure(figsize=(10, 6))  # Adjust figure size for better readability
plt.plot(df['Quarter'], df['Shipments'], marker='o', linestyle='-') # Added markers and lines

plt.xlabel('Quarter', fontsize=12)
plt.ylabel('Shipments (millions of dollars)', fontsize=12)
plt.title('US Household Appliance Shipments (1985-1989)', fontsize=14)
plt.grid(True)  # Add gridlines for better visualization

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45, ha='right')

plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
# prompt: b. Does there appear to be a quarterly pattern? For a closer view of the patterns, zoom in to the range of 3500-5000 on the y-axis.

# Zoom in on the y-axis
plt.figure(figsize=(10, 6))
plt.plot(df['Quarter'], df['Shipments'], marker='o', linestyle='-')
plt.xlabel('Quarter', fontsize=12)
plt.ylabel('Shipments (millions of dollars)', fontsize=12)
plt.title('US Household Appliance Shipments (1985-1989) - Zoomed', fontsize=14)
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.ylim(3500, 5000)  # Set the y-axis limits for zooming
plt.tight_layout()
plt.show()
```
