```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the specified path
df = pd.read_csv('/content/LaptopSalesJanuary2008.csv')

# Calculate the average retail price by store
avg_price_by_store = df.groupby('Store Postcode')['Retail Price'].mean()

# Plot the bar chart
plt.figure(figsize=(10, 6))
avg_price_by_store.plot(kind='bar', color='skyblue')
plt.title('Average Retail Price by Store', fontsize=16)
plt.xlabel('Store Postcode', fontsize=12)
plt.ylabel('Average Retail Price', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()

# Identify the store with the highest and lowest average retail price
max_store = avg_price_by_store.idxmax()
min_store = avg_price_by_store.idxmin()

print(f"Store with the highest average price: {max_store}")
print(f"Store with the lowest average price:{min_store}")
```
